```python
#########################
###### Solution 1.1 #####
#########################


import numpy as np
import matplotlib.pyplot as plt

# parameter range
t = np.linspace(0, 2, 100)  

# Parametric equations
x = t
y = t**2

plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Curve: α(t) = (t, t^2)', color='blue', linewidth=2)
plt.scatter([0, 2], [0, 4], color='red', marker='o', label='Endpoints')
plt.arrow(0, 0, 2, 4, head_width=0.1, head_length=0.1, fc='green', ec='green',linestyle="dotted", label='Direction of Traversal')
plt.title('Solution 1.1: Parametric Curve α(t) = (t, t^2)')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Equal scaling for x and y axes
plt.show()
```

```python
###################################
###### Solution 1.2  ##############
###################################


# parameter range between 0 and 6π
t = np.linspace(0, 6 * np.pi, 100)

# Parametric equations
x = t * np.cos(t)
y = t * np.sin(t)

plt.figure(figsize=(8, 8))
plt.plot(x, y, label='Curve: α(t) = (t cos t, t sin t)', color='blue', linewidth=2)
plt.scatter([0, 6 * np.pi], [0, 0], color='red', marker='o', label='Endpoints')
plt.arrow(0, 0, 20, 20, head_width=0.1, head_length=0.1,color="green",linestyle="dotted", label='Direction of Traversal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution 1.2: Parametric Curve α(t) = (t cos t, t sin t)')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

<b>1.18</b> let $l$ and $m$ be lines in $\mathbb{R}^3$ parametrized by:

$\alpha(t) = a + tv$ and $\beta(t) = b + tw$

respectively, where $a = (a_1, a_2, a_3), v = (v_1, v_2, v_3), b =(b_1, b_2, b_3)$ and $w = (w_1, w_2, w_3)$

<b>(a)</b> Explain why the problem of determining whether $l$ and $m$ intersect in $\mathbb{R}^3$ amounts to solving a system of three equations in two unknowns $t_1$ and $t_2$

<b>(b)</b> Determine whether the lines parametrized by $\alpha(t) = (-1, 0, 1) + t(1, -3, 2)$ and $\beta(t) = (1, 5, 4) + t(-1, 1, 6)$ intersect by writing down the corresponding system of equations, as described in part(a), and solving the system.

<b>(c)</b>Repeat for the lines parametrized by $\alpha(t) = (2, 1, 0) + t(0,0,1)$ and $\beta(t) = (0, 1, 3) + t(1,0,0)$

<b>Solution (a)</b>

Determining if two lines $l$ and $m$ intersect in $\mathbb{R}^n$ requires two distinct parameters $t_1$ and $t_2$ for which $\alpha + tv$ and $\beta + tw$ can coincide.

$\Rightarrow$ given $a = (a_1, a_2, a_3), v = (v_1, v_2, v_3), b =(b_1, b_2, b_3)$ and $w = (w_1, w_2, w_3)$, we need to solve three systems of equations :
$$a_1 + t_1 v_1 = b_1 + t_2 w_1$$
$$a_2 + t_1 v_2 = b_2 + t_2 w_2$$
$$a_3 + t_1 v_3 = b_3 + t_2 w_3$$

$\Rightarrow$ If any values for $t_1$ and $t_2$ satisfy all three equations, solutions exists, otherwise doesn't.


<b>Solution (b)</b>

Using values from  $\alpha(t) = (-1, 0, 1) + t(1, -3, 2)$ and $\beta(t) = (1, 5, 4) + t(-1, 1, 6)$ , we get three system of equations as:

$$-1 + t_1(1) = 1 + t_2(-1)$$
$$0 + t_1(-3) = 5 + t_2(1)$$
$$1 + t_1(2) = 4 + t_2(6)$$

$\Rightarrow$ we have:<br>
$$-1 + t_1 = 1 - t_2\qquad\qquad..eq(1)$$
$$-3t_1 = 5 + t_2\qquad\qquad..eq(2)$$
$$1 + 2t_1 = 4 + t_2\qquad\qquad..eq(3)$$

From, $eq(1)$ we have $t_1 = 2- t_2$<br>
and from $eq(2)$, we get <br>
$$-3(2 - t_2) = 5 + t_2$$
$$\Rightarrow -6 + 3t_2 = 5 + t_2$$
$$\Rightarrow 3t_2 - t_2  = 5 + 6$$
$$\Rightarrow t_2 = \frac{11}{2}$$

Therefore, $t_1 = 2- t_2 = 2 - \frac{11}{2} = -\frac{7}{2}$

But, these values of $t_1$ and $t_2$ when used on $eq(3)$, does not satisfy the equality.

Hence, lines parametrized by $\alpha(t) = (-1, 0, 1) + t(1, -3, 2)$ and $\beta(t) = (1, 5, 4) + t(-1, 1, 6)$ do not intersect.


<b>Solution(c)</b>

Using values from $\alpha(t) = (2, 1, 0) + t(0,0,1)$ and $\beta(t) = (0, 1, 3) + t(1,0,0)$, we get three system of equations:

$$2 + t_1(0) = 0 + t_2(1)$$
$$1 + t_1(0) = 1 + t_2(0)$$
$$0 + t_1(1) = 3 + t_2(0)$$

$\Rightarrow$ solving these equations we have, <br>
$$2 = t_2$$
$$1 = 1$$
$$t_1 = 3$$

There values of $t_1=3$ and $t_2 = 2$ will satisfy all three systems of equations for lines $l$ and $m$. <br>
Therefore, we can say with $\alpha(t) = (2, 1, 0) + t(0,0,1)$ and $\beta(t) = (0, 1, 3) + t(1,0,0)$ parametrization lines $l$ and $m$ intersect.

<b>2.1</b><br>
Let $\alpha(t) = (t^3, 3t^2, 6t)$

<b>(a)</b> Find the velocity, speed, and acceleration<br>
<b>(b)</b> Find the arclength from $t = 0$ to $t = 4$

We know that, <br>
<b>Velocity : $ V(t) = \alpha'(t)$</b> | <b>Acceleration : $\alpha(t) = \alpha''(t)$</b> | <b>Speed : $v(t) = \lVert V(t) \rVert $</b> | 
<b>arclength : $ = \int_a^b v(t)\,dt$</b>

<b>Solution(a)</b><br>

Here, $\frac{d}{dt} t^3 =3t^2$, $\quad \frac{d}{dt}3t^2 = 3\frac{d}{dt}t^2 = 6t\quad$ and $\frac{d}{dt}6t = 6$<br>

Therefore, velocity, $V(t)$ or $\alpha'(t) = (3t^2, 6t, 6)$

Next, speed is $\lVert V(t) \rVert = \lVert \sqrt{(3t^2)^2 + (6t)^2 + 6^2} \rVert$

$\qquad\qquad = \lVert \sqrt{9t^4 + 36t^2 + 36} \rVert$<br>
$\qquad\qquad = \lVert \sqrt{9(t^4 + 4t^2 + 4)} \rVert$<br>
$\qquad\qquad = \lVert \sqrt{9((t^2 + 2)^2)} \rVert$<br>
$\qquad\qquad = \lVert \pm3(t^2 + 2) \rVert$<br>
$\qquad\qquad = 3(t^2 + 2)$

Hence, speed $= 3(t^2 + 2)$

Next, for acceleration, again:

$\frac{d}{dx}3x^2 = 3 \frac{d}{dt}t^2 = 6t\quad, \frac{d}{dt}6t =6\quad$, and $\frac{d}{dt}6 =0$

Therefore, acceleration $\alpha''(t) = (6t, 6, 0)$

<b>Solution(b)</b>

We know that, arclength$ = \int_a^b v(t) dt$

Here, then it would be:<br>
arclength$ = \int_0^4 (3(t^2 + 2))dt$

$\qquad\quad = \int_0^4 (3t^2 + 6) dt$

$\qquad\quad = \int_0^4 3t^2 dt + \int 6 dt$

$\qquad\quad = 3\cdot \frac{t^{2+1}}{2+1} +C + 6t + C$

$\qquad\quad = 3\cdot \frac{t^3}{3} +C + 6t + C$

$\qquad\quad = t^3 + 6t + C$

Now with anitiderivative from $t = 0$  to $t =4$, we have:

$[t^3 + 6t + C]_0^4$

$= (4^3 + 6\cdot 4 + C) - (0^3 + 6 \cdot 0 + C)$

$= 64 + 24 + C - 0 - 0 - C$<br>
$= 88$

Hence, arclength $= 88$

<b>2.2</b><br>
Let $\alpha(t) = (\frac{1}{2}t^2, 2t, \frac{4}{3}t^{3/2})$

<b>(a)</b> Find the velocity, speed, and acceleration<br>
<b>(b)</b> Find the arclength from $t = 2$ to $t = 4$

We know that, <br>
<b>Velocity : $ V(t) = \alpha'(t)$</b> | <b>Acceleration : $\alpha(t) = \alpha''(t)$</b> | <b>Speed : $v(t) = \lVert V(t) \rVert $</b> | 
<b>arclength : $ = \int_a^b v(t)\,dt$</b>

<b>Solution (a)</b>

Here, <br>
$\frac{d}{dt} (\frac{1}{2}t^2) = \frac{1}{2}\cdot \frac{d}{dt}t^2 = \frac{1}{2}\cdot 2t = t$

$\frac{d}{dt}( 2t) = 2 \cdot \frac{d}{dt} t = 2$

$\frac{d}{dt} (\frac{4}{3})(t^{3/2}) = \frac{4}{3}\cdot \frac{3}{2} \cdot\frac{d}{dt} t^{3/2 -1} = 2 t^{1/2}$

Therefore, velocity, $V(t) = \alpha'(t) = (t, 2, 2t^{1/2})$

Again,<br> 
$\frac{d}{dt} t = 1$, $\qquad \frac{d}{dt} 2 = 0$ $\qquad$ and $\frac{d}{dt} (2t^{1/2}) = 2 \cdot \frac{1}{2} \frac{d}{dt} t^{1/2 -1} = t^{-1/2}$

Therefore, acceleration, $\alpha''(t) = (1, 0, t^{-1/2})$ 

Next, we have speed, $v(t) = \lVert V(t) \rVert$ where  $V(t) = (t, 2, 2t^{1/2})$

$\Rightarrow v(t) = \lVert \sqrt{(t)^2 + (2)^2 + (2t^{1/2})^2} \rVert$

$\qquad\quad = \lVert \sqrt{t^2 + 4 + 4t} \rVert$

$\qquad\quad = \lVert \sqrt{(t+2)^2} \rVert$

$\qquad\quad = t+2 $

Hence, speed, $v(t) = t+2$

<b>Solution (b)</b>

We know that, arclength$ = \int_a^b v(t) dt$ 


Here, it would be: <br>
$\int_2^4 (t + 2t) dt$

$ = \int_2^4 t dt + \int_2^4 2\cdot dt$

$= \int_2^4 \frac{t^2}{2} + C + 2t + C$

$ = \int_2^4 \frac{t^2}{2} + 2t + C$

With antiderivative from $t = 2$ to $t = 4$, we can write it as:

$[\frac{t^2}{2} + 2t + C]_2^4$

$ = (\frac{4^2}{2} + 2 \times 4 + C) - (\frac{2^2}{2} + 2 \times 2 + C)$<br>
$ = (8 + 8 + C) - (2 + 4 + C)$<br>
$ = 16- 6$<br>
$ = 10$

Hence, arclength$ = 10$
